```python
import requests
import matplotlib.pyplot as plt
import time


def get_iss_location():
    response = requests.get("http://api.open-notify.org/iss-now.json")
    database = response.json()
    latitude = float(database["iss_position"]["latitude"])
    longitude = float(database["iss_position"]["longitude"])
    return latitude, longitude

datalist = []
for i in range (1,100+1):
    latitude, longitude = get_iss_location()
    datalist.append((latitude, longitude))
    time.sleep(10)
#no longer included the timestamps because the data that is appended automatically goes to the end of list 
    
latitudes = [data[0] for data in datalist]
longitudes = [data[1] for data in datalist]

plt.figure(figsize=(10, 6))
plt.plot(longitudes, latitudes, "-o")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Path of the International Space Station")
plt.grid(True)
plt.show()
```
